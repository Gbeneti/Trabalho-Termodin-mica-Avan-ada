```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')
```

```python
fase = pd.read_csv('benzene-toluene-phase-diagram.txt')
```

```python
x=list(fase.iloc[:,1])
y=list(fase.iloc[:,2])

T=list(fase.iloc[:,0])
```

```python
fig,ax = plt.subplots(1,1)
ax.plot(x, T) 
ax.plot(y,T)
ax.set_ylabel('Temperatura °C')

ax.set_xlabel('Concentração Tolueno')
ax.set_title("Diagrama de fase")
ax.axvline(0.66, linestyle = '--', c = 'orange', label = 'Concentração inicial')
ax.axhline(87.5, linestyle = '--', c = 'r')
ax.axvline(0.834, linestyle = '--', c = '0')
ax.axhline(83.5, linestyle = '--', c = 'r')
ax.axvline(0.93, linestyle = '--', c = '0')
ax.axhline(81.4, linestyle = '--', c = 'r')
ax.axvline(0.97, linestyle = '--', c = '0')
ax.axhline(80.5, linestyle = '--', c = 'r')
ax.axvline(0.99, linestyle = '--', c = '0')
ax.legend()
```

Temperaturas coluna:<br>
$87.5 °C\rightarrow 83.5°C \rightarrow 81.4 °C\rightarrow 80.5°C$ <br>
Concentrações esperadas:<br>
$0.66 \rightarrow 0.834 \rightarrow 0.93 \rightarrow 0.97 \rightarrow 0.99$
